calculating employee count

```python
total_employees = HrData_df.shape[0]
print("Total Employees:", total_employees)
```

calculate attrition count

```python
attrition_count = HrData_df[HrData_df['Attrition'] == 'Yes'].shape[0]
print("Attrition Count:", attrition_count)
```

calculate attrition rate

```python
attrition_rate = (attrition_count / total_employees) * 100
print("Attrition Rate: {:.2f}%".format(attrition_rate))
```

Active employees

```python
active_employees = HrData_df[HrData_df['Attrition'] == 'No'].shape[0]
print("Active Employees:", active_employees)
```

Average age

```python
average_age = HrData_df['Age'].mean()
print("Average Age of Employees:", average_age)
```

No. of employees by Age group


```python


print(HrData_df['Age'].isnull().sum())

HrData_df = HrData_df.dropna(subset=['Age'])

bins = [ 0,20, 30, 40, 50, 60, 70]

labels = ['<20','20-29', '30-39', '40-49', '50-59', '60-69']

HrData_df['Age Group'] = pd.cut(HrData_df['Age'], bins=bins, labels=labels, right=False)

age_group_counts = HrData_df['Age Group'].value_counts().sort_index()
print(age_group_counts)

plt.figure(figsize=(5, 6))
age_group_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Employees by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.show()
```

Department wise Attrition

```python

attrition_by_department = HrData_df[HrData_df['Attrition'] == 'Yes'].groupby('Department').size()

plt.figure(figsize=(8, 8))
plt.pie(attrition_by_department, labels=attrition_by_department.index, autopct='%1.1f%%', startangle=140)
plt.title('Attrition by Department')
plt.show()
```

Job satisfaction rating

```python
plt.figure(figsize=(12, 6))
HrData_df.groupby('Job Role')['Job Satisfaction'].mean().plot(kind='bar', color='skyblue')
plt.title('Job Role vs. Average Job Satisfaction')
plt.xlabel('Job Role')
plt.ylabel('Average Job Satisfaction')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
HrData_df.boxplot(column='Job Satisfaction', by='Job Role', vert=False)
plt.title('Job Role vs. Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Job Role')
plt.show()
```

Education Field wise Attrition

```python
attrition_by_education_field = HrData_df[HrData_df['Attrition'] == 'Yes'].groupby('Education Field').size()


plt.figure(figsize=(10, 6))
attrition_by_education_field.plot(kind='bar', color='skyblue')
plt.title('Attrition by Education Field')
plt.xlabel('Education Field')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(attrition_by_education_field, labels=attrition_by_education_field.index, autopct='%1.1f%%', startangle=140)
plt.title('Attrition by Education Field')
plt.show()
```
